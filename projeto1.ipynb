#### **Fontes de dados e técnicas utilizadas**

#### **Scheduler e logs**

O processo foi configurado para ser executado de forma periódica através do GitHub Actions, recorrendo a workflows definidos na pasta `.github/workflows/`. Estes workflows funcionam como alternativa a um scheduler local, permitindo automatizar a recolha de dados ao longo do tempo.

Na pasta `.github/workflows/`, foram definidos ficheiros `.yml` separados para automatizar os diferentes processos de scraping desenvolvidos pelos elementos do grupo. Estes workflows permitem executar periodicamente os scrapers através de expressões cron e também podem ser iniciados manualmente através da opção `workflow_dispatch`.

Durante a recolha de dados, foram registadas mensagens em ficheiros de logs, o que permitiu acompanhar o número de artigos recolhidos, os erros encontrados e as execuções em que não foram identificados novos dados.

Como cada elemento do grupo desenvolveu uma parte separada do scraping, os logs podem ser mantidos em ficheiros independentes. Esta abordagem permite acompanhar individualmente a execução de cada recolha, facilitando a identificação de erros e a verificação dos dados obtidos por cada scraper. Assim, não é necessário juntar todos os logs num único ficheiro, uma vez que cada ficheiro documenta a execução da respetiva parte do processo.

```python
from pathlib import Path
import json

dataset_path = Path("data/processed/dataset_final.json")
log_dir = Path("logs")
workflow_dir = Path(".github/workflows")
workflow_files = sorted(workflow_dir.glob("*.yml"))

print("Dataset existe:", dataset_path.exists())
print("Logs existem:", log_dir.exists() and any(log_dir.glob("*.log")))
print("Workflows encontrados:", len(workflow_files))
for workflow_file in workflow_files:
    print("-", workflow_file)
```

#### **Carregamento do ficheiro JSON em memória**

De seguida, o ficheiro JSON final é carregado em memória. Este ficheiro reúne todos os registos que foram recolhidos e normalizados.

```python
with dataset_path.open("r", encoding="utf-8") as f:
    noticias = json.load(f)

# A variável news fica definida para ser reutilizada na Tarefa 2.
news = noticias

print("Total de registos no dataset final:", len(noticias))
print("Campos disponíveis no primeiro registo:", list(noticias[0].keys()))
```

#### **Visualização do JSON final**

Uma vez que o ficheiro JSON final contém um número elevado de registos, é apresentado abaixo apenas um excerto do dataset final. O ficheiro completo encontra-se em data/processed/dataset_final.json.

```python
print(json.dumps(noticias[:2], indent=2, ensure_ascii=False))
```

#### **Resumo do dataset recolhido**

Para verificar a dimensão e a diversidade dos dados recolhidos, é realizada uma contagem dos registos por domínio de origem.

```python
from collections import Counter

fontes = Counter(noticia.get("source", {}).get("domain", "desconhecido") for noticia in noticias)

for fonte, total in fontes.most_common():
    print(f"{fonte}: {total} registos")
```

#### **Percorrer o conteúdo de 5 registos**

Por fim, são percorridos e apresentados 5 registos do dataset, conforme solicitado no enunciado. Para cada registo, são mostrados o título, a fonte, o URL, a data de publicação e uma parte do conteúdo textual.

```python
for i, noticia in enumerate(noticias[:5], start=1):
    print(f"--- Registo {i} ---")
    print("ID:", noticia.get("id"))
    print("Título:", noticia.get("title"))
    print("Fonte:", noticia.get("source", {}).get("name"))
    print("Domínio:", noticia.get("source", {}).get("domain"))
    print("URL:", noticia.get("url"))
    print("Publicado em:", noticia.get("published_at"))
    print("Recolhido em:", noticia.get("collected_at"))
    print("Autor:", noticia.get("author") or "N/A")
    print("Conteúdo:", noticia.get("content", "")[:700])
    print()
```

#### **Verificação dos logs**

A célula seguinte apresenta as últimas linhas dos ficheiros de logs existentes, onde é possível acompanhar o desenrolar do processo de scraping.

```python
log_files = sorted(log_dir.glob("*.log"))

if log_files:
    for log_file in log_files:
        print(f"\n--- {log_file} ---")

        with log_file.open("r", encoding="utf-8", errors="replace") as f:
            linhas = f.readlines()

        for linha in linhas[-5:]:
            print(linha.strip())
else:
    print("Nenhum ficheiro de logs encontrado.")
```
