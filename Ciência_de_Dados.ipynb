### Armazenamento no Redis com Docker


Para este projeto, o Redis foi executado num **contentor Docker**, o que permite uma configuração simples, isolada e reprodutível em qualquer máquina, sem necessidade de instalação directa no sistema operativo.

#### Ficheiro `docker-compose.yml`

O ficheiro abaixo define o serviço Redis. A imagem utilizada é a `redis/redis-stack`, que inclui o Redis e módulos adicionais como o **RediSearch**.

````

  redis:

      timeout: 90s

````

#### Ligação ao Redis em Python

Após o container estar em execução, a ligação é estabelecida com o parâmetro `decode_responses=True`, que garante que as respostas do Redis são devolvidas como strings Python em vez de bytes.

`````python
import redis

# Conectar ao Redis a correr no Docker
r = redis.Redis(
    host='localhost',
    port=6379,
    decode_responses=True
)
`````

#### Carregamento do Dataset

O ficheiro `dataset_final.json` é carregado em memória para que os artigos possam ser indexados no Redis. Este ficheiro contém todos os registos normalizados recolhidos na Tarefa 1.

`````python
import json

# Carregar dataset
with open('dataset_final.json', 'r', encoding='utf-8') as f:
    news = json.load(f)

print(f"Notícias carregadas: {len(news)}")
`````

### Indexação das Notícias

Os artigos foram indexados no Redis utilizando a estrutura de dados **Hash**. Um Hash no Redis permite modelar cada notícia como um conjunto de campos nomeados sob uma única chave

Cada notícia é armazenada com uma chave única no formato `news:{id}`, onde `{id}` é o identificador único de cada artigo. Os campos indexados incluem o título, conteúdo, fonte, domínio, URL, autor e datas de publicação e recolha.

`````python
for x in news:
    chave = f"news:{x['id']}"
    r.hset(chave, mapping={
        'id'          : x['id'],
        'title'       : x['title'],
        'content'     : x['content'],
        'source'      : x['source']['name'],
        'domain'      : x['source']['domain'],
        'url'         : x['url'],
        'author'      : x.get('author') or 'N/A',
        'published_at': x.get('published_at') or 'N/A',
        'collected_at': x.get('collected_at') or 'N/A',
    })

print(f"Total de chaves no Redis: {r.dbsize()}")
`````

### Recuperação de Informação — Queries

Para demonstrar a recuperação de informação a partir do Redis, foram executadas 4 queries sobre os dados indexados. Como as notícias estão armazenadas como Hashes, a recuperação é feita com `r.hgetall(key)`, que devolve todos os campos de um artigo como dicionário Python.

`````python
# Query 1 - Ver dados de uma Notícia específica

print ("=== Query 1 - Ver dados de uma notícia específica ===")

noticia_teste = r.hgetall("news:e58568c642986e4d81872a60b27d3078")

if noticia_teste:
    print("Notícia Teste")
    print(f"Título  : {noticia_teste.get('title', '')}")
    print(f"Fonte   : {noticia_teste.get('source', 'N/A')}")
    print(f"Autor   : {noticia_teste.get('author', 'N/A')}")
    print(f"URL     : {noticia_teste.get('url', '')}")


# Query 2 - Ranking de autores

print("\n")
print ("=== Query 2 - Ranking de autores ===")

# Apagar ranking anterior para não acumular
r.delete("ranking_autores")
for n in news:
    autor = n.get('author') or "Desconhecido"
    r.zincrby("ranking_autores", 1, autor)

# Query: Top 5 autores
autores = r.zrevrange("ranking_autores", 0, 4, withscores=True)

for autor, qtd in autores:
    print(f"Autor: {autor} -> Total: {int(qtd)}")


# Query 3 - Notícias que mencionam "machine learning"

print("\n")
print ("=== Query 3 - Noticias que mencionam 'machine learning' ===")

termo = "machine learning"
resultados = []

for key in r.scan_iter("news:*"):
    news   = r.hgetall(key)
    texto = (news.get('title','') + " " + news.get('content','')).lower()
    if termo in texto:
        resultados.append((news.get('title',''), news.get('source',''), news.get('published_at','')))

# Ordenar por data
resultados.sort(key=lambda x: x[2], reverse=True)

for titulo, fonte, data in resultados[:5]:
    print(f"Data  : {data}")
    print(f"Título: {titulo}")
    print(f"Fonte : {fonte}")
    print(f"URL : {news.get('url', '')}")
    print()

# Query 4 - Artigos da fonte "sapo" que mencionam "ia"

print("\n")
print ("=== Query 4 - Notícias da fonte 'sapo' que mencionam 'ia' ===")

resultados = []

for key in r.scan_iter("news:*"):

    new = r.hgetall(key)
    
    fonte = new.get('source', '')
    texto = (new.get('title', '') + " " + new.get('content', '')).lower()
    
    if fonte == "sapo" and "ia" in texto:
        resultados.append(new)
     # Top 5
    if len(resultados) == 5:
        break

for artigo in resultados:
    print(artigo)
    
`````
